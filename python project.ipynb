# Univariate Analysis - one variable 
# Histograam

```python
df.describe()
```

```python
bins=np.arange(df['math_score'].min(),df['math_score'].max()+30,30)
counts,bins,patches=plt.hist(df['math_score'],bins=bins)
print(counts) #[91. 41. 32. 28. 12.  3.  1.]
print(bins) #[  0.  30.  60.  90. 120. 150. 180. 210.]
print(bins[1]) # index of the bins
bin_width=bins[1]-bins[0]
print(len(counts))

for i in range(len(counts)): # len(counts) = 7
    plt.text(bins[i]+bin_width/2,counts[i],str(int(counts[i])),ha='center',va='bottom',fontsize=12)
    plt.title("Histogram on Maths Score")
plt.xlabel("Math score") # X-axis how much is the score from 0 to last run
plt.ylabel("Frequency") # Y-axis how frequently the person scored
plt.xticks(bins,rotation=45) # X-axis value edits - eg: no '30' in x-axis is rotated in 45 degree
plt.show()
```

```python
Q1=df['math_score'].quantile(0.25)
Q3=df['math_score'].quantile(0.75)
minimum=df['math_score'].min()
plt.boxplot(df['math_score'])
plt.text(x=1.1,y = df['math_score'].min(),s = f"min: {minimum}")
plt.text(x=1.1,y = df.math_score.quantile(0.25),s = f"Q1 -> {Q1}")
plt.text(x=1.1,y = df['math_score'].median(),s = f"median: (Q2)")
plt.text(x=1.1,y = df.math_score.quantile(0.75),s = f"Q3 -> {Q3}")
plt.text(x=1.1,y = df['math_score'].max(),s = f"max")
#the graph tittle and axes labels
plt.title('Boxplot of Math Score')
plt.ylabel('math score')
```

```python
# Sorted countplot:
plt.figure(figsize=(8,6))
sns.countplot(data=df,y="parental_level_of_education",order=df['parental_level_of_education'].value_counts().index,palette="coolwarm")
plt.title("Frequency of parental level of education ",fontsize=16)
plt.xlabel("count",fontsize=12)
plt.ylabel("parental level of education Mode",fontsize=12)
plt.show()
```

```python
# Customization:
sns.set(style="whitegrid")
# Scatterplot: Runs vs. Balls Faced
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="reading_score", y="writing_score", hue="gender" ,palette="plasma")
# other palette -'plasma', 'coolwarm', 'cividis',"viridis"
plt.title("writing score vs. reading score	", fontsize=16)
plt.xlabel("reading score	", fontsize=12)
plt.ylabel("writing score", fontsize=12)
plt.legend(title="gender", loc='upper left')
plt.show()
```

```python
sns.barplot(data=df,x="gender",y="math_score",palette="cividis") 
plt.show()
```

```python
parental_level_of_education_counts=df["parental_level_of_education"].value_counts()
print(parental_level_of_education_counts.index)
print(parental_level_of_education_counts.values)
```

```python
explode=[0,0,0,0,0,0.30] # Explosion level outside the piechart
plt.figure(figsize=(10, 10)) # set the plot size
#colors=['red', 'blue', 'green', 'yellow', 'black', 'purple', 'Lightblue']
plt.pie(parental_level_of_education_counts, labels=parental_level_of_education_counts.index,
autopct='%0.2f%%', startangle=140, # Rotation angle - 0-360 degree
colors=sns.color_palette("viridis"), explode=explode)
# can also use colors instead of color palette
plt.title("parental level of education Types Proportion", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="race_ethnicity", y="math_score", palette="coolwarm", hue="gender")
plt.title("Math score by race ethnicity", fontsize=16)
plt.xlabel("race ethnicity", fontsize=12)
plt.ylabel("Math score", fontsize=12)
plt.show()
```

```python
numeric_data = df.select_dtypes (include=[np.number])
corr_matrix = numeric_data.corr()
corr_matrix
```

```python
# Plot the heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(corr_matrix, annot=True, cmap="cividis", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Variables", fontsize=14)
plt.show()
```
